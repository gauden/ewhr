##Â Temporary notebook to test the database of figure properties

```python
import plot
```

```python
def get_props(notebook_no):
    try:
        props = plot.FigureProperties(notebook_no)
        return props.rec
    except KeyError:
        return 'No figure with that ID ({}) found.'.format(notebook_no)
```

```python
for notebook_no in [1,'02',6,'7',12,18,24,'nonesuch']:
    print('Getting properties for notebook #: {}'.format(notebook_no))
    print(get_props(notebook_no))
    print('-' * 50)
```
