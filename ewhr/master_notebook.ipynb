```python
from IPython.display import HTML
```

```python
DEBUG = False  # when debugging is on, only run two notebooks
```

```python
def report(nbs, rec):
    """Produce user-friendly report on notebook result."""
    report_tpl = """
    <h1>Fig. {fig_no}. {title}</h1>
    <div style="margin: 15">{thumb}</div>
    <p><a href="../fig_{fig_no}.html">View local copy</a></p>
    <hr>
    """
    
    error_tpl = """
    <h1>Notebook {} generated no valid plot.</h1>
    <hr>
    """
    
    try:
        return report_tpl.format(fig_no=rec.specs['fig_no'],
                                 title=rec.specs['title'],
                                 thumb=rec.specs['thumb'])
    except:
        return error_tpl.format(nbs)
```

# Notebook to run all others

This notebook runs all others in the folder. Its main use: to update the static website with new template features.

```python
output = ''

nb_list = [4, 1] if DEBUG else [4, 1, 2, 3, 5, 6,
                                7, 9, 8, 11, 10, 12,
                                13, 15, 14, 16, 18, 17]

for nb in nb_list:
    rec = None  # clear away remnants of any previous notebook
    nbs = 'notebook_{}.ipynb'.format(str(nb).zfill(2))
    print('\nProcessing {}'.format(nbs))
    %run $nbs
    output += report(nbs, rec)
    
HTML(output)
```
